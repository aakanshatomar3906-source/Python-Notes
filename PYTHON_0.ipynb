CONDITIONAL STATEMENTS:-

1. IF-ELSE

```python
age=18
if age>=18:
  print("You are eligible")
else:
  print("You are not eligible to vote.")
```

```python
a=int(input("Enter a number:"))
if a>0:
  print("Positive")
if a==0:
  print("Zero")
else:
  print("Negative")
```

2. ELIF STATEMENT

```python
score=85
if score>=90:
  print("Grade A")
elif score>=80:
  print("Grade B")
else:
  print("Grade C")
```

3. NESTED IF


```python
month='december'
time='afternoon'

if (month=='december'):
  if(time=='afternoon'):
    print("let's go")
  else: print("1 condition doesnot  met")
else:
  print("2 condition doesnot met")
```

```python
num=10
if num>0:
  if num%2==0:
    print("positive and even")
  else:
    print("positive and odd")
```

```python
num=int(input("Enter a number"))
if num%2==0:
  if num%3==0:
    print("divisible by 2 and 3")
  else: print("divisible by 2 but not with 3")
else:
  print("doesnot divisible by 2 ")
```

LOOPS


1. FOR LOOP

```python
fruits=['apple','banana','cheery']
for fruit in fruits:
  print(fruits)
```

```python
fruits=['apple','banana','cheery']
for fruit in fruits:
  print(fruit)
```

```python
for i in range(10,0,-2):
  print(i)
```

```python
for i in range(10,-1,-2):
  print(i)
```

```python
for i in range(10,0,2):
  print(i)
```
